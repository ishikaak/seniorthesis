```python
import os
import pytesseract
from PIL import Image
from jiwer import wer

# Configure Tesseract OCR path

# Define input folder for images and ground truth file
input_folder = '/content/drive/My Drive/washingtondb-v1.0/data/line_images_normalized'
ground_truth_file = '/content/drive/My Drive/washingtondb-v1.0/ground_truth/transcription.txt'

# Function to clean and preprocess ground truth lines
def preprocess_ground_truth(line):
    # Remove line numbers (assuming they are at the start and followed by a space)
    line = line.split(' ', 1)[-1] if ' ' in line else line
    # Replace "|" with spaces
    line = line.replace('|', ' ')
    # Remove dashes
    line = line.replace('-', '')
    return line.strip()

# Load and preprocess ground truth data
with open(ground_truth_file, "r", encoding="utf-8") as f:
    ground_truth_lines = [preprocess_ground_truth(line.strip()) for line in f.readlines()]

# Ensure ground truth data matches the number of images
image_files = sorted(
    [file for file in os.listdir(input_folder) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp'))]
)
if len(image_files) != len(ground_truth_lines):
    raise ValueError("Number of images does not match the number of lines in the ground truth file.")

# Process each image
for image_file, ground_truth in zip(image_files, ground_truth_lines):
    image_path = os.path.join(input_folder, image_file)

    # OCR with pytesseract
    extracted_text = pytesseract.image_to_string(Image.open(image_path))

    # Calculate and print WER
    error_rate = wer(ground_truth, extracted_text)
    print(f"{image_file} - WER: {error_rate:.2%}")
    print(f"Extracted: {extracted_text}")
    print(f"Ground Truth: {ground_truth}\n")
```
